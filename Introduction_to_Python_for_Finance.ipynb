```python
# def square(x):

def square(x):
    x_squared = x**2
    return x_squared

def cube(x):
    return x**3

    
square(2)
cube(3)
```

```python
def fib(x):
    fib = [0,1]
    for i in range(1, x):
      fib.append(fib[-1] + fib[-2])
    return fib[-1]

def fib2(x):
  if x <= 1:
       return x
  else:
       return(fib2(x-1) + fib2(x-2))

for i in range(11):
  print(fib(i))

for i in range(11):
  print(fib2(i))
```

```python
def factorial(n):
    #    """
    # write a function that returns factorial of n : n * (n-1) * (n-2) * (n-3) ...
    # """
  result = 1
  for i in range(1, n + 1):
    result = result * i  
  return result
      
def fac2(x):
  if x == 1:
    return 1
  elif x == 2:
    return 2
  else :
    return x*fac2(x-1)

print(factorial(6))
fac2(6)
  
    
```

```python
def coef_binomial(k,n):
    # """
    # write a function that returns the binomal coefficient (k,n)
    # """

    return factorial(n) / (factorial(k)*factorial(n-k))

coef_binomial(5,10)
```

```python
def pascal_triangle(n):
    # """
    # write a function that prints the pascal triangle up to line n
    
    # 1
    # 1 1
    # 1 2 1
    # 1 3 3 1
    # 1 4 6 4 1
    # 1 5 10 10 5 1
    
    # """

    for i in range(n):
      for j in range(i +1):
        print(int(coef_binomial(j, i)), end = ' ')
      print()

pascal_triangle(10)
```
