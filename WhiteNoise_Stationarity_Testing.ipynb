```python
import pandas as pd
import statsmodels.tsa.stattools as stattools
import statsmodels.api as stats_lb
import numpy as np



def stationarity_test(file_name, column_name): 
    df = pd.read_csv(file_name)
    df["log_return"] = np.log(df["spot"]/df["spot"].shift(1))
    series = df['log_return'].dropna().head(5000).values

    adf_result = stattools.adfuller(series)
    kpss_result = stattools.kpss(series, regression='c')
    lb_test = stats_lb.stats.acorr_ljungbox(series)
    lb_pvalue = lb_test["lb_pvalue"].tail(1).values
    
    print("test for:", file_name)

    print("Ljung-Box test p-value:", lb_pvalue)
    if lb_pvalue < 0.05:
        print("The p-value is low, indicating strong evidence against the null hypothesis of no autocorrelation in the residuals.")
    else:
        print("The p-value is high, indicating weak evidence against the null hypothesis of no autocorrelation in the residuals.")

    print('ADF Statistic: %f' % adf_result[0])
    print('p-value: %f' % adf_result[1])
    if adf_result[1] < 0.05:
        print("The p-value is low, indicating strong evidence against the null hypothesis of the time series being non-stationary.")
    else:
        print("The p-value is high, indicating weak evidence against the null hypothesis of the time series being non-stationary.")

    print('KPSS Statistic: %f' % kpss_result[0])
    print('p-value: %f' % kpss_result[1])
    if kpss_result[1] < 0.05:
        print("The p-value is low, indicating strong evidence against the null hypothesis of the time series being stationary.")
    else:
        print("The p-value is high, indicating weak evidence against the null hypothesis of the time series being stationary.")

```

```python
csv_ls = ['boom-300.csv','boom-500.csv', 'boom-1000.csv', 'jump-10.csv','jump-25.csv', 'jump-50.csv','jump-75.csv','jump-100.csv','vol-10-1s.csv','vol-10-2s.csv','vol-100-1s.csv','vol-100-2s.csv','vol-25-1s.csv', 'vol-25-2s.csv' , 'vol-50-1s.csv', 'vol-50-2s.csv' , 'vol-75-1s.csv', 'vol-75-2s.csv']


for file in csv_ls:
    stationarity_test(file,"spot")
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import datetime
import matplotlib.dates as mdates
```

```python
def plot_acf_pacf(file_name):

    df = pd.read_csv(file_name)
    df["ts"] = pd.to_datetime(df["ts"])
    df = df.set_index("ts")

    df1 = df.resample("H").last()
    df1["log_diff"] = np.log(df1["spot"]/df1["spot"].shift(1))
    df1 = df1[df1["log_diff"].notna()].reset_index()

    plt.rcParams.update({'figure.figsize':(10,5), 'figure.dpi':120})
    fig, axes = plt.subplots(2, 3,sharex=True)
    axes[0, 0].plot(df1["spot"])
    axes[0, 0].set_title('Spot Series')
    axes[0, 0].set(xlim=(-1,20))
    plot_acf(df1["spot"], ax=axes[0, 1], lags = 20)
    plot_pacf(df1["spot"], ax=axes[0, 2], lags = 20)

    axes[1, 0].plot(df1["log_diff"])
    axes[1, 0].set_title('1st Order Differenced Series')
    axes[1, 0].set(xlim=(-1, 20))
    plot_acf(df1["log_diff"].dropna(), ax=axes[1, 1], lags = 20)
    plot_pacf(df1["log_diff"].dropna(), ax=axes[1, 2], lags = 20)

    plt.show()
```

```python
csv_ls = ['boom-300.csv','boom-500.csv', 'boom-1000.csv', 'jump-10.csv','jump-25.csv', 'jump-50.csv','jump-75.csv','jump-100.csv','vol-10-1s.csv','vol-10-2s.csv','vol-100-1s.csv','vol-100-2s.csv','vol-25-1s.csv', 'vol-25-2s.csv' , 'vol-50-1s.csv', 'vol-50-2s.csv' , 'vol-75-1s.csv', 'vol-75-2s.csv']

white_noise = np.random.normal(0, 1, 1000)

print("/-----------","white noise","-----------/")

lags = 40
plt.rcParams.update({'figure.figsize':(10,5), 'figure.dpi':120})
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(white_noise)
axes[0, 0].set_title('Spot Series')
axes[0, 0].set(xlim=(-1,lags))
plot_acf(white_noise, ax=axes[0, 1], lags = lags);

axes[0, 1].set_ylabel("autocorrelation")
axes[0, 1].set_xlabel("lags")
plot_pacf(white_noise, ax=axes[1, 1], lags = lags);
axes[1, 1].set_ylabel("partial autocorrelation")
axes[1, 1].set_xlabel("lags")

fig.delaxes(axes[1, 0])
fig.tight_layout()
plt.show()

for file in csv_ls:
    print("-----------",file,"-----------")
    plot_acf_pacf(file)
```
