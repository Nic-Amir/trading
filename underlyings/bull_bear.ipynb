```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
dt = 1/86400
x0 = 100
d= -35
np.random.seed(1)

sigma = 1.75

x = np.exp(
    (-d - sigma ** 2 / 2) * dt + sigma * np.random.normal(0, np.sqrt(dt), size=(1, n)).T
)
x = np.vstack([np.ones(1), x])
x = x0 * x.cumprod(axis=0)

plt.plot(x)
# plt.legend(sigma)
plt.xlabel("$t$")
plt.ylabel("$x$")
plt.title(
    "Realizations of Geometric Brownian Motion with positive drift and 175% volatility"
)
plt.show()
```
