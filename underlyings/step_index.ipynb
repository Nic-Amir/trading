```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10000
dt = 1/(365*24*60*60)
x0 = 100

np.random.seed(1)
random = np.random.uniform(0,1, size=(n))
direction = np.where(random>0.5,1,-1)

x = direction
x = x0 + x.cumsum(axis=0)

plt.plot(x)
plt.xlabel("$t$")
plt.ylabel("$x$")
plt.title(
    "Simple Step Index Example"
)
plt.show()
```
